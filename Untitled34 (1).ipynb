```python
import pandas as pd

df = pd.read_csv("data/customer_support_tickets.csv")

df.head()
```

```python
text_column = None

for col in df.columns:
    if df[col].dtype == "object":
        text_column = col
        break

print("Using column:", text_column)
```

```python
tickets = df[text_column].dropna().astype(str).tolist()

print("Total tickets:", len(tickets))
```

```python
labels = [
    "Network Issue",
    "Login Issue",
    "Billing Issue",
    "Technical Bug",
    "Account Issue",
    "Subscription Issue",
    "General Inquiry"
]
```

```python
from transformers import pipeline

classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)
```

```python
def get_top_tags(text):
    result = classifier(text, labels)

    top_3 = sorted(
        zip(result["labels"], result["scores"]),
        key=lambda x: x[1],
        reverse=True
    )[:3]

    return top_3
```

```python
few_shot_prompt = """
Classify support tickets into:
Network Issue, Login Issue, Billing Issue, Technical Bug, Account Issue, Subscription Issue.

Examples:
Ticket: Internet not working → Network Issue
Ticket: Cannot login → Login Issue
Ticket: Payment failed → Billing Issue

Now classify:
"""

def few_shot_classify(text):
    result = classifier(few_shot_prompt + text, labels)

    top_3 = sorted(
        zip(result["labels"], result["scores"]),
        key=lambda x: x[1],
        reverse=True
    )[:3]

    return top_3
```

```python
for i in range(3):
    print("\nTicket:", tickets[i])

    print("Zero-shot:", get_top_tags(tickets[i]))
    print("Few-shot:", few_shot_classify(tickets[i]))
```

```python
results = []

for text in tickets[:50]:
    tags = get_top_tags(text)

    results.append({
        "ticket": text,
        "tag_1": tags[0][0],
        "tag_2": tags[1][0],
        "tag_3": tags[2][0]
    })

output_df = pd.DataFrame(results)

output_df.head()
```

```python
output_df.to_csv("auto_tagged_tickets.csv", index=False)

print("CSV saved successfully")
```

```python
import matplotlib.pyplot as plt

top_labels = output_df["tag_1"].value_counts()

plt.figure(figsize=(8,5))
plt.bar(top_labels.index, top_labels.values)

plt.title("Support Ticket Category Distribution")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```
